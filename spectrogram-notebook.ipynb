```python
import pickle
import librosa
import matplotlib.pyplot as plt
import librosa.display as disp
import numpy as np
import random
```

```python
genre_list = ["blues", "classical", "country", "disco", "hiphop", "jazz", "metal", "pop", "reggae", "rock"]
with open("pickled-data/train_spect_music_data.pkl", "rb") as f:
    music_data = pickle.load(f)

audio = music_data["audio"]
num_of_tracks = len(audio)
print(music_data["labels"][3000])
```

```python
for i in range(0,5):
    track_id = random.randint(0, num_of_tracks)
    plt.figure()
    disp.specshow(audio[track_id], x_axis='s', y_axis='mel', hop_length=256, fmin = 0, fmax=22050/2)
    title_str = "Spectrogram of " + genre_list[music_data["labels"][track_id]] + " genre"
    plt.title(title_str)
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")
```
